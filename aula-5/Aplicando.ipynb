### Importando os dados

```r
inventory <- read.csv("Estatística Preditiva - Aplicando 5.csv")
head(inventory)
```

### Calculando a correlação

Percebemos que `Mês` se correlaciona fortemente com as outras variáveis. Da mesma todas as outras variáveis entre si. 

```r
corr_matrix <- cor(inventory)
print(corr_matrix)
```

### Regressão linear

Utilizando `X1` e `X2` como variáveis preditoras, calculamos um plano que melhor prediz `Y`. Percebemos, através dos `p-values` dos coeficientes que `X1` é uma boa variável preditora de `Y` porém `X2` nem tanto. Por outro lado `R-squared` é de 0.94, o que indica que o plano preditor se aproxima bem dos dados de treinamento. Visualizando os dados abaixo, notamos que ambas as variáveis se correlacionam positivamente fortemente com `Y`.

```r
reg <- lm(Y ~ X1 + X2, data = inventory)

summary(reg)
```
