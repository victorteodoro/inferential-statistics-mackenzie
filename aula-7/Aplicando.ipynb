```r
days <- 1:12
vehicles <- c(188, 272, 170, 256, 204, 248, 256, 233, 178, 175, 168, 166)

df <- data.frame(days = days, vehicles = vehicles)
head(df)

```

```r
library(ggplot2)
library(quantmod)

moving_avg <- TTR::SMA(df$vehicles, n = 3)
print(round(moving_avg, 2))
```

```r
ggplot(df, aes(x = days)) +
  geom_line(aes(y = vehicles, color = "Série original"), group = 1) +
  geom_line(aes(y = moving_avg, color = "Média móvel"),group = 1) +
  theme(legend.position = "top") +
  labs(color = "Legenda:")
```
